```python
from langchain.llms import HuggingFacePipeline
import auto_gptq
```

```python
from transformers import AutoTokenizer, pipeline, logging
from auto_gptq import AutoGPTQForCausalLM, BaseQuantizeConfig

quantized_model_dir = "TheBloke/vicuna-7B-v1.5-GPTQ"

tokenizer = AutoTokenizer.from_pretrained(quantized_model_dir, use_fast=False)

def get_config(has_desc_act):
    return BaseQuantizeConfig(
        bits=4,  # quantize model to 4-bit
        group_size=128,  # it is recommended to set the value to 128
        desc_act=has_desc_act
    )

def get_model(model_base, triton, model_has_desc_act):
    if model_has_desc_act:
        model_suffix="latest.act-order"
    else:
        model_suffix="compat.no-act-order"
    return AutoGPTQForCausalLM.from_quantized(quantized_model_dir, use_safetensors=True, device="cuda:0", use_triton=triton, quantize_config=get_config(model_has_desc_act)) #model_basename=f"{model_base}.{model_suffix}"
```

```python
model = get_model("vicuna-7B-v1.5-GPTQ", triton=False, model_has_desc_act=False)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,
    temperature=0.001,
    top_p=0.95,
    repetition_penalty=1.15
)
```

```python
from langchain import PromptTemplate, HuggingFaceHub, LLMChain

vicuna_template="""You are a spokesperson for a left-leaning political advocacy group with liberal views. You are engaging in a debate with a representative from a conservative think tank. Your goal is to present well-reasoned arguments based on liberal principles. Express your viewpoint clearly and provide evidence to support your stance. If the other representative makes compelling points, be open to adjusting your opinions. Your responses should be concise, with a maximum length of 200 words.
1. Begin by introducing the core liberal principle related to the topic.
2. Use Socratic questioning to prompt the conservative representative's perspective.
3. Provide evidence to support your argument and encourage counterpoints.
4. Be prepared to adjust your stance based on the strengths of the opposing arguments.

The topic of discussion is {question}.

Your response:


 """


prompt1 = PromptTemplate(template=vicuna_template, input_variables=["question"])
hf = HuggingFacePipeline(pipeline=pipe)
```

```python
llm_chain1 = LLMChain(prompt=prompt1,
                     llm=hf)
```

```python
template2 = """You are a spokesperson for a right-leaning political advocacy group with conservative views. You are engaging in a debate with a representative from a liberal non-profit organization. Your objective is to articulate conservative principles and provide evidence to support your arguments. Maintain a clear stance and be open to adjusting your opinions if the other representative presents compelling points. Keep your responses concise, with a maximum length of 200 words.
1. Start by outlining the core conservative principle related to the topic.
2. Utilize Socratic questioning to explore the liberal representative's perspective.
3. Support your argument with evidence and encourage counterarguments.
4. Demonstrate flexibility by adjusting your opinions in response to persuasive opposing points.

The topic of discussion is {question}.

Your response:


"""

prompt2 = PromptTemplate(template=template2, input_variables=["opinion"])
```

```python
llm_chain2=LLMChain(prompt=prompt2,
                     llm=hf)
```

```python
with open('responses3.txt', 'a', encoding='UTF8') as f:
    seed_prompt = """Are monopolies good for the economy?"""
    answer1 = llm_chain1.run(seed_prompt) #opinion
    print(answer1)
    print("-------------")
    f.write("Agent1: "+ answer1+"\n")
    for i in range(10):
        answer2 = llm_chain2.run(answer1) #question
        print(answer2)
        print("-------------")
        f.write("Agent2: "+answer2+"\n")
        f.write('\n')
        answer1 = llm_chain1.run(answer2) # opinion
        print(answer1)
        print("-------------")
        f.write("Agent1: "+answer1+"\n")
        f.write('\n')
```
